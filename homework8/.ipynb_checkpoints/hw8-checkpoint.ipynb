```python
# The different columns represent the abbrevieated name of the state, 
#the latitude of said state, the longitude of said state, and 
#the name of the state.

#The first state is Alabame

#The last state is Wyoming

#There are a total of 52 entries

#There are more than 50 because it follows the standard of the first index being zero.
```

```python
median_latitude=df['latitude'].median() #The value is 39.4355155
median_latitude
```

```python
median_longitude=df['longitude'].median() #The value is -89.093198
median_longitude
```

```python
df['Absolute Value to Median Latitude']=(df['latitude']-median_latitude).abs()
df
```

```python
df['Absolute Value to Median Longitude']=(df['longitude']-median_longitude).abs()
df
```

```python
df.shape #I now have 52 rows and 6 columns
```

```python
df.loc[df['Absolute Value to Median Latitude'].idxmin(), 'name'] #using df.loc lets me find the index
#of the row and use that to find the state name.
print(f"The state closest to the median latitude is {df.loc[df['Absolute Value to Median Latitude'].idxmin(), 'name']}, by {df['Absolute Value to Median Latitude'].min() : .2f} degrees!")
print(f"The state farthest to the median latitude is {df.loc[df['Absolute Value to Median Latitude'].idxmax(), 'name']}, by {df['Absolute Value to Median Latitude'].max() : .2f} degrees!")
print(f"The state closest to the median longitude is {df.loc[df['Absolute Value to Median Longitude'].idxmin(), 'name']}, by {df['Absolute Value to Median Longitude'].min() : .2f} degrees!")
print(f"The state farthest to the median longitude is {df.loc[df['Absolute Value to Median Longitude'].idxmax(), 'name']}, by {df['Absolute Value to Median Longitude'].max() : .2f} degrees!")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("taxis")
```

```python
df.shape #there are 6433 rows and 14 columns
```

```python
df.head()
```

```python
df.tail()
```

```python
#pickup represents the time and date the passanger was picked up. 
#dropoff represents the time and date the passanger was dropped off. 
#passengers refers to the number of people in the taxi.
#distance refers to how far the ride for the passenger was.
#fare represents the cost for their ride was.
#tip refers to the tip amount the passenger gave to the driver.
#tolls I believe refers to the amount of tolls that were passed by, this probably also accounts to the fare cost.
#total refers to the total cost of the ride, includes fare and tip.
#color refers to the color of the taxi.
#pickup_zone refers to the place of pickup.
#dropoff_zone refers to the place where the passenger had been dropped off at.
#pickup_borough refers to the borough in NY where the passenger was picked up at.
#dropoff_borough refers to the borough in NY where the passenger was dropped off at.
```

```python
fig,axes= plt.subplots(2,2, figsize=(12,8))

sns.scatterplot(data=df, x="fare", y="tip", color="orange", hue="tip", ax=axes[0,0])
#used "hue" in the upper left scatter plot which functions as a legend
sns.scatterplot(data=df, x="passengers", y="distance", ax=axes[0,1], color="green")
sns.histplot(data=df, x="pickup_borough",ax=axes[1,0], color="yellow")
sns.boxplot(data=df, x="passengers", ax=axes[1,1], color="red")

axes[0,0].set_title("Tip Vs Fare")
axes[0,0].set_xlabel("Fare ($)")
axes[0,0].set_ylabel("Tip ($)")

axes[0,1].set_title("Passengers Vs Distance")
axes[0,1].set_xlabel("Amount of Passengers")
axes[0,1].set_ylabel("Distance Traveled (mi)")

axes[1,0].set_title("Which Borough Brings In More Money?")
axes[1,0].set_xlabel("Borough of Pick Up")
axes[1,0].set_ylabel("Amount of Instances")

axes[1,1].set_title("Passengers Boxplot")
axes[1,1].set_xlabel("Amount of Passengers")

plt.tight_layout()
plt.show()
```

```python
#Upper left:
#a)This subplot represents the relationship between fare prices and the tip given. I wanted to see if there was
#a relationship between the more expensive a ride was, the more a person would be willing to tip more.
#b)This dataset taught me that it seems to be that the more expensive a ride was, the bigger the tip.

#Upper Right:
#a)This represents the amount of passengers and how far they traveled. I chose this because I thought that 
#maybe these two would show me that if there's more people in a taxi, they would probably choose to travel farther.
#b)I learned that my idea wasn't correct. If anything, it looks like people traveled farther alone. But other than that, 
#there does not seem to be any trends. So the amount of people in a taxi does not indicate how far they travel.

#Lower Left:
#a)This set represents how many pick-ups happened at each borough. I decided on this
#coloumn because I wanted to see which borough was more popular, and what this would mean revenue wise.
#b)I learned that Manhattan was the spot to be for the taxi industry. Which I assumed it would mean this was because it was the 
#most populated, but I was way wrong! It's actually Brooklyn which the bigger population. Thus, after a few searches, 
#I learned that Manhattan is the heart of NY, and this is where all the activites, such as buisnesses, tourist spots, etc. are at.
#So it makes sense that this place would see the most activity in terms of taxis.

#Lower Right:
#a) This represents how passengers vary by taxi, informing of averages, medians, and outliers. I wanted to see how passangers vary and which is most seen. 
#b) I learned that as in the Passengers Vs. Distance plot, 1-2 people per taxi is the popular choice. Where it is then rare to see taxis with passengers up to 4,5, and 6 people.
```
