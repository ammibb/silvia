```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_sine_wave(h,A,w):
    sinwh=np.sin(w*h)
    y=A*sinwh
    return y
x=np.linspace(0,2*np.pi,1000)
amplitude=np.array([0.5,1,1.5,2,2.5])
frequency=np.array([1,2,3,4,5])
```

```python
fig, axes= plt.subplots(1,3, figsize=(25,5))
#varying amplitude using make_sine_wave
for A in amplitude: #so i get the value
    y=make_sine_wave(x,A,1)
    axes[0].plot(x,y, label=f"A = {A}")
axes[0].set_xlabel("X-Values")
axes[0].set_ylabel("Sine Values")
axes[0].set_title("Varying Amplitude")

for f in frequency: #varying frequecy
    y=make_sine_wave(x,1,f) #this makes an array for f, which allows this to be plotted
    axes[1].plot(x,y, label=f"F = {f}")
axes[1].set_title("Varying Frequency")
axes[1].set_xlabel("X-Values")
axes[1].set_ylabel("Sine Values")

for A,f in zip(amplitude, frequency): #looping through both, simultaneously
    y= make_sine_wave(x,A,f)
    axes[2].plot(x,y, label=f"A = {A}, F = {f}")
axes[2].set_title("Varying Both Frequency and Amplitude")
axes[2].set_xlabel("X-Values")
axes[2].set_ylabel("Sine Values")
#use .legend() on each individual suplot
axes[0].legend()
axes[1].legend()
axes[2].legend()
plt.show()
```
