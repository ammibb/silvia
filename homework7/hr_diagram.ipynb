```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data= np.loadtxt("../../course_assignments/homework7/star_data.txt", skiprows=1)
```

```python
np.shape(data) #1st col, magnitude (brightness) of the star in units of magnitude
#2nd col, the temp of the star in units of Kelvin
#there are 2,642 rows
```

```python
magnitude=data[:,0] #The lower the magnitude, the brighter a star
temperature=data[:,1]
```

```python
print(np.min(magnitude))
```

```python
print(np.max(magnitude))
```

```python
np.argmin(magnitude) #The min magnitude corresponds to the 1989th star
```

```python
np.argmax(magnitude)  #The max magnitude corresponds to the 265th star
```

```python
#The brightest star is the 1989th (1988th index) star, whereas the dimmest is the 265th star.
```

```python
log_magnitude=np.log(magnitude)
log_temperature=np.log(temperature) #using plt.figure() for a single figure:
plt.figure(figsize=(8,6))
plt.scatter(log_temperature, log_magnitude)
plt.gca().invert_xaxis() #gets the current x-axis and flips it 
plt.gca().invert_yaxis() #gets the current y-axis and flips it
#assigning specific colors
bri_dim=[1988, 264]
hot_cold=[2613, 1923]
temp_colors=["Green", "Orange"]
mag_colors=["Black", "Purple"]
#Brightest Star
plt.scatter(log_temperature[1988], log_magnitude[1988], color="Black", marker="s") #so i can use this again
#but the idea is that since they work in pairs, the 1988th value will match with the 1988th value in log_temperature as well
#Dimmest Star
plt.scatter(log_temperature[264], log_magnitude[264], color="Purple", marker="h")
#Hottest Star
plt.scatter(log_temperature[2613], log_magnitude[2613], color="Green", marker="D")
#Coldest Star
plt.scatter(log_temperature[1923], log_magnitude[1923], color="Orange", marker="*")

plt.title("Hertzsprung-Russel Diagram")
plt.xlabel("Temperature (log K)")
plt.ylabel("Magnitude")
plt.plot()
```
