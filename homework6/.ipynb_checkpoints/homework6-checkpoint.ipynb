```python
#4 Prime Star Systems
```

```python
starmags=np.array([[2,3,5],[14,6,8],[11,13,17],[7,10,13]])
```

```python
#I create a function which checks if a number is prime
def prime_number(number): 
    if number<2:
        return False
    for i in range(2, int(number**0.5) +1): #plus 1 since range is start (included), stop (excluded). Int is also neeeded since range only accepts integers
        if number%i==0:
            return False #the function is stopped if a factor is found, otherwise, it returns True
    else:
        return True
```

```python
#using np.vectorize is much more efficient, this works on each value in the arrray
#and i can create my own using this method as well.
#Then, using np.any, it'll sends back the rows which have at least one true value
```

```python
prime_matrix=np.vectorize(prime_number)
valuesofprime=prime_matrix(starmags) #this saves which values are prime and which are not
valuesofprime
```

```python
#finally, this function keeps the values of the rows which have at least one prime, or not
leftover=np.any(valuesofprime, axis=1) 
#this line then has the array starmags keep the rows where the boolean is true and their respective values.
prime_rows=starmags[leftover]
prime_rows 
```

```python
#5.2
zero8by8=np.zeros((8,8))  #shape gives me a tuple, with shape[0] being the amount of rows, and shape[2] being the amount of columns
for i in range(0,zero8by8.shape[0], 2): #so i want the rows, this targets the odd numbered rows
    for j in range(0, zero8by8.shape[1], 2):#this gives me the amount of values in each odd row
        zero8by8[i,j]=1
zero8by8
#Now i'll modify the even rows
for k in range(1,zero8by8.shape[0],2):
    for l in range(1, zero8by8.shape[0], 2):
        zero8by8[k,l]=1
zero8by8
```

```python
def reverse_checkerboard(zero8by8):
    reversed_order=np.fliplr(zero8by8) #this function allows me to flip the rows from left to right!
    return reversed_order
print(reverse_checkerboard(zero8by8))
```

```python
#6 The Expanding Universe
```

```python
def expansion(word, num): #i can take the index, and to that, add spaces
    #for each letter, add num amount of spaces:
    length=len(word)
    new_word=""
    amnt_of_spaces=" " * num
    for p in range(0, length-1):
        new_word+= word[p]+ amnt_of_spaces
    new_word+=word[length-1] #these 2 lines ensure there are no spaces at the end
    return new_word

expansion("kittykat", 4)
```

```python
#7 Second-Brightest Star
```

```python
#2d array where each row is a star cluster, return array where second brightest returned
#from each row
#so, create a new array, loop though the shape, add the 2nd brightest to the array
def secondDimmest(assort):
    arr_2ndDimmest=[]
    sortedstars=np.sort(assort)#this arranges each value in each row from least to greatest
    place=sortedstars.shape[1]-2 #gives me the amount of indexes in a row, -2 for targeting the 2nd dimmest
    #Looping through each row
    for v in range(sortedstars.shape[0]):
        arr_2ndDimmest.append(sortedstars[v,place])
    return np.array(arr_2ndDimmest)
```

```python
stars=np.random.randint(500,2000, (5,5))
stars
```

```python
secondDimmest(stars)
```
